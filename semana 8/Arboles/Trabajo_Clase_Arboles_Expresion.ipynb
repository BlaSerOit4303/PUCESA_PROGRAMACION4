## Ejercicio 1
**Expresión a resolver:** `(7 + 2) * (5 - 3)`

### a. Construcción del Árbol

```python
# Escribe aquí el código para construir el árbol
# Importamos la librería matplotlib para generar gráficos (plt es el alias común).
import matplotlib.pyplot as plt

# Importamos la librería networkx para trabajar con grafos dirigidos (nx es el alias común).
import networkx as nx

# Creamos un nuevo grafo dirigido (DiGraph significa "Directed Graph", es decir, con flechas).
G = nx.DiGraph()

# Definimos las conexiones (aristas) entre los nodos. Cada tupla representa una conexión de padre a hijo.
edges = [
    ('*', '+'),  # El nodo raíz '*' tiene como hijo izquierdo el operador '+'
    ('*', '-'),  # El nodo raíz '*' tiene como hijo derecho el operador '-'
    ('+', '7'),  # El nodo '+' tiene como hijo izquierdo el operando '7'
    ('+', '2'),  # El nodo '+' tiene como hijo derecho el operando '2'
    ('-', '5'),  # El nodo '-' tiene como hijo izquierdo el operando '5'
    ('-', '3')   # El nodo '-' tiene como hijo derecho el operando '3'
]

# Agregamos todas las aristas al grafo G. NetworkX creará automáticamente los nodos implicados.
G.add_edges_from(edges)

# Definimos las posiciones en el plano (x, y) para que el árbol se visualice ordenado y jerárquico.
pos = {
    '*': (0, 2),      # La raíz está en el nivel más alto
    '+': (-1, 1),     # Hijo izquierdo de '*'
    '-': (1, 1),      # Hijo derecho de '*'
    '7': (-1.5, 0),   # Hijo izquierdo de '+'
    '2': (-0.5, 0),   # Hijo derecho de '+'
    '5': (0.5, 0),    # Hijo izquierdo de '-'
    '3': (1.5, 0)     # Hijo derecho de '-'
}

```

### b. Visualización con `networkx`

```python
# Escribe aquí el código para graficar el árbol
# Creamos una nueva figura de tamaño 8x5 pulgadas
plt.figure(figsize=(8, 5))

# Dibujamos el grafo:
# - G: el grafo a dibujar
# - pos: las posiciones definidas para cada nodo
# - with_labels=True: muestra los nombres (etiquetas) de los nodos
# - node_color='lightblue': color de los nodos
# - node_size=2000: tamaño de los nodos
# - font_size=14, font_weight='bold': tamaño y grosor del texto
# - arrows=True: muestra flechas indicando dirección
nx.draw(
    G, pos,
    with_labels=True,
    node_color='lightblue',
    node_size=2000,
    font_size=14,
    font_weight='bold',
    arrows=True
)

# Título del gráfico
plt.title("Árbol de expresión: (7 + 2) * (5 - 3)")

# Muestra el gráfico en pantalla
plt.show()
```

### c. Evaluación de la Expresión

```python
# Escribe aquí el código para evaluar la expresión
# Evaluación de la expresión ((7 + 2) * (5 - 3)) usando árbol binario

class Nodo:
    def __init__(self, valor):
        self.valor = valor  # Guarda el valor del nodo (puede ser un número o un operador)
        self.izq = None  # Referencia al hijo izquierdo del nodo
        self.der = None  # Referencia al hijo derecho del nodo

def construir_arbol():
    raiz = Nodo('*')  # Nodo raíz con el operador '*'
    raiz.izq = Nodo('+')  # Hijo izquierdo de la raíz con el operador '+'
    raiz.der = Nodo('-')  # Hijo derecho de la raíz con el operador '-'
    raiz.izq.izq = Nodo(7)  # Hijo izquierdo del nodo '+' con valor 7
    raiz.izq.der = Nodo(2)  # Hijo derecho del nodo '+' con valor 2
    raiz.der.izq = Nodo(5)  # Hijo izquierdo del nodo '-' con valor 5
    raiz.der.der = Nodo(3)  # Hijo derecho del nodo '-' con valor 3
    return raiz  # Devuelve el nodo raíz del árbol construido

# Función para evaluar el árbol de expresión de manera recursiva
def evaluar(nodo):
    if isinstance(nodo.valor, int):   # Si es un número, se retorna directamente
        return nodo.valor
    izq = evaluar(nodo.izq)           # Evaluar subárbol izquierdo
    der = evaluar(nodo.der)           # Evaluar subárbol derecho
    if nodo.valor == '+':             # Suma
        return izq + der
    elif nodo.valor == '-':           # Resta
        return izq - der
    elif nodo.valor == '*':           # Multiplicación
        return izq * der
    elif nodo.valor == '/':           # División
        return izq / der

# Construimos y evaluamos el árbol
arbol = construir_arbol()  # Llama a la función para construir el árbol y guarda la raíz
resultado = evaluar(arbol)  # Evalúa la expresión representada por el árbol
resultado  # Muestra el resultado final de la evaluación
```

```python
# Escribe aquí el código para evaluar la expresión
class Nodo:
    def __init__(self, valor):
        self.valor = valor  # Asigna el valor al nodo (puede ser un número o un operador)
        self.izq = None  # Referencia al hijo izquierdo del nodo
        self.der = None  # Referencia al hijo derecho del nodo

# Construcción del árbol binario para la expresión.
def construir_arbol():
    raiz = Nodo('-')  # Nodo raíz con el operador de resta '-'
    raiz.izq = Nodo(9)  # Hijo izquierdo de la raíz con el valor 9
    raiz.der = Nodo('+')  # Hijo derecho de la raíz con el operador '+'
    raiz.der.der = Nodo('/')  # Hijo derecho del '+' con el operador '/'
    raiz.der.der.izq = Nodo(8)  # Hijo izquierdo del '/' con el valor 8
    raiz.der.der.der = Nodo(4)  # Hijo derecho del '/' con el valor 4
    raiz.der.izq = Nodo("*")  # Hijo izquierdo del '+' con el operador '*'
    raiz.der.izq.izq = Nodo(3)  # Hijo izquierdo del '*' con el valor 3
    raiz.der.izq.der = Nodo(2)  # Hijo derecho del '*' con el valor 2
    return raiz  # Devuelve la raíz del árbol construido

# Función para evaluar recursivamente el árbol.
def evaluar(nodo):
    if isinstance(nodo.valor, int):  # Si el nodo es un número
        return nodo.valor  # Retorna directamente el valor numérico
    izq = evaluar(nodo.izq)  # Evalúa recursivamente el subárbol izquierdo
    der = evaluar(nodo.der)  # Evalúa recursivamente el subárbol derecho
    if nodo.valor == '+':  # Si el operador es '+'
        return izq + der  # Retorna la suma de los subresultados
    elif nodo.valor == '-':  # Si el operador es '-'
        return izq - der  # Retorna la resta de los subresultados
    elif nodo.valor == '*':  # Si el operador es '*'
        return izq * der  # Retorna la multiplicación de los subresultados
    elif nodo.valor == '/':  # Si el operador es '/'
        return izq / der  # Retorna la división de los subresultados

# Se construye el árbol y se evalúa.
arbol = construir_arbol()  # Se construye el árbol de la expresión y se guarda en 'arbol'
resultado = evaluar(arbol)  # Se evalúa el árbol usando la función recursiva
resultado  # Muestra el resultado final de la evaluación
```

### a. Construcción del Árbol

```python
# Escribe aquí el código para construir el árbol
# Importamos la librería matplotlib para generar gráficos (plt es el alias común).
import matplotlib.pyplot as plt

# Importamos la librería networkx para trabajar con grafos dirigidos (nx es el alias común).
import networkx as nx

# Creamos un nuevo grafo dirigido (DiGraph significa "Directed Graph", es decir, con flechas).
G = nx.DiGraph()

# Definimos las conexiones (aristas) entre los nodos. Cada tupla representa una conexión de padre a hijo.
edges = [
    ('/', '+.1'),       # '/' tiene como hijo izquierdo '+.1'
    ('/', '-.1'),       # '/' tiene como hijo derecho '-.1'
    ('+.1', '5'),       # '+.1' tiene como hijo izquierdo el número 5
    ('+.1', '*'),       # '+.1' tiene como hijo derecho '*'
    ('*', '2.a'),       # '*' tiene como hijo izquierdo el número 2
    ('*', '-.2'),       # '*' tiene como hijo derecho '-.2'
    ('-.2', '8'),       # '-.2' tiene como hijo izquierdo el número 8
    ('-.2', '3'),       # '-.2' tiene como hijo derecho el número 3
    ('-.1', '6'),       # '-.1' tiene como hijo izquierdo el número 6
    ('-.1', '+.2'),     # '-.1' tiene como hijo derecho '+.2'
    ('+.2', '1'),       # '+.2' tiene como hijo izquierdo el número 1
    ('+.2', '2.b'),     # '+.2' tiene como hijo derecho el número 2
]
# Agregamos todas las aristas al grafo G. NetworkX creará automáticamente los nodos implicados.
G.add_edges_from(edges)

# Definimos las posiciones en el plano (x, y) para que el árbol se visualice ordenado y jerárquico.
pos = {
    '/': (0, 2),            # Nodo raíz
    '+.1': (-1.5, 1),       # Hijo izquierdo de '/'
    '-.1': (1.5, 1),        # Hijo derecho de '/'
    '5': (-2, 0),           # Hijo izquierdo de '+.1'
    '*': (-1, 0),           # Hijo derecho de '+.1'
    '2.a': (-1.5, -1),      # Hijo izquierdo de '*'
    '-.2': (-0.5, -1),      # Hijo derecho de '*'
    '8': (-1, -2),          # Hijo izquierdo de '-.2'
    '3': (0, -2),           # Hijo derecho de '-.2'
    '6': (1, 0),            # Hijo izquierdo de '-.1'
    '+.2': (2, 0),          # Hijo derecho de '-.1'
    '1': (1.5, -1),         # Hijo izquierdo de '+.2'
    '2.b': (2.5, -1),       # Hijo derecho de '+.2'
}
# Etiquetas de los nodos para visualización
labels = {
    '/': '/',
    '+.1': '+',
    '+.2': '+',
    '-.1': '-',
    '-.2': '-',
    '*': '*',
    '5': '5',
    '8': '8',
    '3': '3',
    '6': '6',
    '1': '1',
    '2.a': '2',
    '2.b': '2',
}
```

### b. Visualización con `networkx`

```python
# Escribe aquí el código para graficar el árbol
# Dibujamos el árbol
plt.figure(figsize=(10, 5))
nx.draw(
    G, pos,
    with_labels=True,
    labels=labels,
    node_color='lightblue',
    node_size=2000,
    font_size=14,
    font_weight='bold',
    arrows=True
)
plt.title("Árbol de expresión: (5 + (2 * (8 - 3))) / (6 - (1 + 2))")
plt.show()
```

### c. Evaluación de la Expresión

```python
# Escribe aquí el código para evaluar la expresión
class Nodo:
    def __init__(self, valor):
        self.valor = valor    # Valor del nodo
        self.izq = None       # Hijo izquierdo
        self.der = None       # Hijo derecho

# Construcción del árbol
def construir_arbol():
    raiz = Nodo('/')                     # Nodo raíz '/'
    raiz.izq = Nodo('+')                # Parte izquierda: suma
    raiz.der = Nodo('-')                # Parte derecha: resta

    raiz.izq.izq = Nodo(5)              # Hijo izquierdo de '+' es 5
    raiz.izq.der = Nodo('*')            # Hijo derecho de '+' es '*'
    raiz.izq.der.izq = Nodo(2)          # Hijo izquierdo de '*' es 2
    raiz.izq.der.der = Nodo('-')        # Hijo derecho de '*' es '-'
    raiz.izq.der.der.izq = Nodo(8)      # Hijo izquierdo de '-' es 8
    raiz.izq.der.der.der = Nodo(3)      # Hijo derecho de '-' es 3

    raiz.der.izq = Nodo(6)              # Hijo izquierdo de '-' es 6
    raiz.der.der = Nodo('+')            # Hijo derecho de '-' es '+'
    raiz.der.der.izq = Nodo(1)          # Hijo izquierdo de '+' es 1
    raiz.der.der.der = Nodo(2)          # Hijo derecho de '+' es 2

    return raiz

# Evaluación recursiva del árbol
def evaluar(nodo):
    if isinstance(nodo.valor, int):     # Si es número, retornamos
        return nodo.valor
    izq = evaluar(nodo.izq)             # Evaluamos hijo izquierdo
    der = evaluar(nodo.der)             # Evaluamos hijo derecho
    if nodo.valor == '+':               # Suma
        return izq + der
    elif nodo.valor == '-':             # Resta
        return izq - der
    elif nodo.valor == '*':             # Multiplicación
        return izq * der
    elif nodo.valor == '/':             # División
        return izq / der

# Construimos el árbol y evaluamos
arbol = construir_arbol()
resultado = evaluar(arbol)  # Resultado final de la evaluación
resultado
```
